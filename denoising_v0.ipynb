```python
def prepareImages(dataset,fig):#replace fig by img_nr
    file_list=os.listdir(dataset).sort(reverse=True)
    #fig=file_list[img_nr]
    print("preparing image: "+dataset+"/"+fig)
    fileList=os.listdir(dataset)
    #for fig in fileList:
    img=image.load_img(dataset+"/"+fig, color_mode = "grayscale")
    x=image.img_to_array(img)
    firstRow=x[[0,]]
    lastRow=x[[(x.shape[0]-1)],:]
    x=np.concatenate((firstRow,firstRow,x,lastRow,lastRow), axis=0)
    firstCol=x[:,[0]]
    lastCol=x[:,[x.shape[1]-1]]
    x=np.concatenate((firstCol,firstCol,x,lastCol,lastCol), axis=1)
    n=0
    X=np.zeros(((x.shape[0]-4)*(x.shape[1]-4),25))
    #X=np.zeros((226800,5,5,1))
    for i in range(0,x.shape[0]-4):
                   for j in range(0,x.shape[1]-4):
                                  tmp=x[i:i+5,j:j+5]
                                  X[n]=tmp.flatten()
                                  n+=1                            
    X=X.astype('float32')/255                              
    return X
```

```python
train_images=prepareImages('train',"2.png")
test_images=prepareImages('test',"4.png")
print(train_images.dtype)
print(train_images.shape)
print(test_images.shape)
```

### read in train label

```python
#yList=os.listdir('train_cleaned')
#fig=yList[0]
def prepareY(Ydata,fig):
    print("preparing image: "+Ydata+"/"+fig)
    fileList=os.listdir(Ydata)
    cleanImg=image.load_img(Ydata+"/"+fig, color_mode = "grayscale")
    y=image.img_to_array(cleanImg)
    y=y.flatten()
    Y=np.transpose(y)
    return Y    
```

```python
Y=prepareY('train_cleaned',"2.png")
```

```python
print(Y.shape)
print(Y.dtype)
print(Y[1:10])
print(max(Y))
print(min(Y))
#n, bins, patches = plt.hist(Y, 256, density=1, facecolor='green', alpha=0.75)
#plt.show()
```

```python
from keras.utils import to_categorical
```

```python
train_labels=to_categorical(Y,256)
print(train_labels.shape)
```

### Create Validation Set here

```python
val_train=train_images[:30000]
partial_train_images=train_images[30000:]
val_label=train_labels[:30000]
partial_train_label=train_labels[30000:]
```

## Model

```python
batch_size = 128
num_classes = 256
epochs = 10
input_shape = (train_images.shape[1],)
```

```python
from keras import models
from keras import layers
model=models.Sequential()
#model.add(layers.Dense(16,activation='relu', input_shape=(10000,)))
#model.add(layers.Dense(16, activation='relu'))
#model.add(layers.Dense(1,activation='sigmoid'))
model.add(layers.Dense(512,activation='relu',input_shape=input_shape))
model.add(layers.Dense(num_classes, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy'])
history=model.fit(partial_train_images,
                partial_train_label,
                epochs=20,
                batch_size=512,
                validation_data=(val_train,val_label))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
history_dict=history.history
loss_values=history_dict['loss']
val_loss_values=history_dict['val_loss']

epochs=range(1,len(loss_values)+1)
plt.plot(epochs,loss_values, 'bo',
        label='Verlust Training')
plt.plot(epochs,val_loss_values, 'b',
        label='Verlust Validierung')
plt.title('Wert der Verlustfunktionen Training/Validierung')
plt.xlabel('Epochen')
plt.ylabel('Wert der Verlustfunktionen')
plt.legend()
plt.show()
```

```python
plt.clf()
acc=history_dict['acc']
val_acc=history_dict['val_acc']

plt.plot(epochs, acc, 'bo', label='Training')
plt.plot(epochs, val_acc, 'b', label='Validierung')
plt.title('Korrektklassifizierungsrate Training/Validierung')
plt.xlabel('Epochen')
plt.ylabel('Korrektklassifizierungsrate')
plt.legend()
plt.show()
```
